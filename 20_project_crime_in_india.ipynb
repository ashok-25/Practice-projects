```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# What is the major reason people being kidnapped in each and every state

```python
df1=pd.read_csv('39_Specific_purpose_of_kidnapping_and_abduction.csv')
df1.head()
```

```python
df1.tail()
```

```python
df1.dtypes
```

```python
df1.describe()
```

```python
df1.nunique()
```

```python
df1.isnull().sum()
```

```python
plt.figure(figsize=(15,8))
a=sns.barplot(x='Area_Name',y='K_A_Cases_Reported',hue='Group_Name',data=df1,palette='rocket')
a.set_xticklabels(labels=df1['Area_Name'].unique(),rotation=90)
```

We can see that in all Areas reason for kidnap is Adoption and Begging.

```python
corri=df1.corr()
corri
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corri,annot=True,cmap='YlGnBu')
```
